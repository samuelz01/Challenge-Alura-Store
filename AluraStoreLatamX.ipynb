### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# Calcular ingresos totales de cada tienda
ingreso_tienda1 = tienda["Precio"].sum()
ingreso_tienda2 = tienda2["Precio"].sum()
ingreso_tienda3 = tienda3["Precio"].sum()
ingreso_tienda4 = tienda4["Precio"].sum()

# Guardar los resultados en un diccionario
ingresos = {
    "Tienda 1": ingreso_tienda1,
    "Tienda 2": ingreso_tienda2,
    "Tienda 3": ingreso_tienda3,
    "Tienda 4": ingreso_tienda4
}

# Convertir a DataFrame para verlo mejor
df_ingresos = pd.DataFrame(list(ingresos.items()), columns=["Tienda", "Ingreso Total"])

print("Ingresos totales por tienda:")
print(df_ingresos)
```

# 2. Ventas por categoría

```python
# ================================
# Cantidad de productos vendidos por categoría en cada tienda
# ================================

# Tienda 1
categorias_tienda1 = tienda.groupby("Categoría del Producto")["Producto"].count().sort_values(ascending=False)
print("Categorías más populares - Tienda 1:\n", categorias_tienda1, "\n")

# Tienda 2
categorias_tienda2 = tienda2.groupby("Categoría del Producto")["Producto"].count().sort_values(ascending=False)
print("Categorías más populares - Tienda 2:\n", categorias_tienda2, "\n")

# Tienda 3
categorias_tienda3 = tienda3.groupby("Categoría del Producto")["Producto"].count().sort_values(ascending=False)
print("Categorías más populares - Tienda 3:\n", categorias_tienda3, "\n")

# Tienda 4
categorias_tienda4 = tienda4.groupby("Categoría del Producto")["Producto"].count().sort_values(ascending=False)
print("Categorías más populares - Tienda 4:\n", categorias_tienda4, "\n")
```

# 3. Calificación promedio de la tienda


```python
import matplotlib.pyplot as plt
```

```python
# ================================
# Calcular calificación promedio por tienda
# ================================

calif_tienda1 = tienda["Calificación"].mean()
calif_tienda2 = tienda2["Calificación"].mean()
calif_tienda3 = tienda3["Calificación"].mean()
calif_tienda4 = tienda4["Calificación"].mean()

# Guardamos en un DataFrame para verlo ordenado
calificaciones = {
    "Tienda 1": calif_tienda1,
    "Tienda 2": calif_tienda2,
    "Tienda 3": calif_tienda3,
    "Tienda 4": calif_tienda4
}

df_calificaciones = pd.DataFrame(list(calificaciones.items()), columns=["Tienda", "Calificación Promedio"])
print(df_calificaciones)

# ================================
# Visualización
# ================================
df_calificaciones.set_index("Tienda").plot(kind="bar", legend=False, title="Calificación Promedio por Tienda", color="skyblue")
plt.ylabel("Calificación promedio")
plt.show()
```

# 4. Productos más y menos vendidos

```python
# ================================
# Productos más y menos vendidos por tienda
# ================================

import matplotlib.pyplot as plt

datasets = [(tienda, "Tienda 1"), (tienda2, "Tienda 2"), (tienda3, "Tienda 3"), (tienda4, "Tienda 4")]

for df, nombre in datasets:
    ventas_productos = df.groupby("Producto")["Precio"].count().sort_values(ascending=False)

    # Producto más vendido
    top_productos = ventas_productos.head(5)
    # Producto menos vendido
    bottom_productos = ventas_productos.tail(5)

    print(f"\n{nombre} - Productos más vendidos:\n", top_productos)
    print(f"\n{nombre} - Productos menos vendidos:\n", bottom_productos)

    # Graficar más vendidos
    top_productos.plot(kind="bar", title=f"{nombre} - Top 5 productos más vendidos", color="green")
    plt.ylabel("Cantidad de ventas")
    plt.show()

    # Graficar menos vendidos
    bottom_productos.plot(kind="bar", title=f"{nombre} - Top 5 productos menos vendidos", color="red")
    plt.ylabel("Cantidad de ventas")
    plt.show()
```

# 5. Envío promedio por tienda

```python
# ================================
# Costo de envío promedio por tienda
# ================================

envio_tienda1 = tienda["Costo de envío"].mean()
envio_tienda2 = tienda2["Costo de envío"].mean()
envio_tienda3 = tienda3["Costo de envío"].mean()
envio_tienda4 = tienda4["Costo de envío"].mean()

# Guardamos en un DataFrame para verlo ordenado
envios = {
    "Tienda 1": envio_tienda1,
    "Tienda 2": envio_tienda2,
    "Tienda 3": envio_tienda3,
    "Tienda 4": envio_tienda4
}

df_envios = pd.DataFrame(list(envios.items()), columns=["Tienda", "Costo de Envío Promedio"])
print(df_envios)

# ================================
# Visualización
# ================================
import matplotlib.pyplot as plt

df_envios.set_index("Tienda").plot(kind="bar", legend=False, color="orange", title="Costo de envío promedio por tienda")
plt.ylabel("Costo promedio de envío")
plt.show()
```

```python
import matplotlib.pyplot as plt

# ================================
# 1. Ingresos totales por tienda (Barras)
# ================================
df_ingresos.set_index("Tienda").plot(
    kind="bar",
    legend=False,
    color="skyblue",
    title="Ingresos Totales por Tienda"
)
plt.ylabel("Ingresos ($)")
plt.show()


# ================================
# 2. Distribución de categorías en Tienda 1 (Gráfico de pastel)
# ================================
categorias_tienda1 = tienda.groupby("Categoría del Producto")["Producto"].count()

categorias_tienda1.plot(
    kind="pie",
    autopct='%1.1f%%',
    title="Distribución de categorías - Tienda 1",
    figsize=(6,6)
)
plt.ylabel("")
plt.show()


# ================================
# 3. Relación entre precio y calificación (Dispersión)
# ================================
plt.scatter(tienda["Precio"], tienda["Calificación"], alpha=0.5, c="green")
plt.title("Relación entre Precio y Calificación - Tienda 1")
plt.xlabel("Precio")
plt.ylabel("Calificación")
plt.show()
```

## Introducción  
El objetivo de este análisis fue identificar la tienda más adecuada para que el **Sr. Juan** realice la venta de sus productos.  
Para ello, se analizaron diferentes factores clave:  

- Ingresos totales por tienda.  
- Categorías de productos más y menos vendidas.  
- Calificaciones promedio de los clientes.  
- Productos más y menos vendidos.  
- Costos de envío promedio.  

La combinación de estadísticas y visualizaciones permitió obtener una visión integral del rendimiento de cada tienda, evaluando fortalezas y debilidades.  

---

## Desarrollo del Análisis  

### 💰 Ingresos Totales  
- **Tienda 1** lidera con **1.150.880.000**, seguida de cerca por **Tienda 2 (1.116.344.000)** y **Tienda 3 (1.098.020.000)**.  
- **Tienda 4** ocupa el último lugar con **1.038.376.000**.  

### 🛍️ Categorías de Productos  
En todas las tiendas predominan las categorías de **Muebles y Electrónicos**, seguidas de **Juguetes y Electrodomésticos**.  
Esto muestra que hay un patrón claro en las preferencias de compra de los clientes.  

### ⭐ Calificaciones Promedio  
- **Tienda 3** → **4.05**, la más alta.  
- **Tienda 2** → **4.04**, muy cercana a Tienda 3.  
- **Tienda 4** → **3.99**, promedio aceptable.  
- **Tienda 1** → **3.97**, la más baja.  

Esto indica que los clientes están más satisfechos en **Tienda 3**.  

### 📦 Productos más y menos vendidos  
- **Tienda 1**: destaca en electrónicos (TV LED UHD 4K, Microondas).  
- **Tienda 2**: fuerte en libros digitales y electrodomésticos.  
- **Tienda 3**: sobresale en muebles grandes (camas, mesas, sets).  
- **Tienda 4**: se orienta a muebles y artículos para el hogar.  

Cada tienda tiene un mix de productos distinto, pero las categorías más demandadas son consistentes.  

### 🚚 Costos de Envío Promedio  
- **Tienda 4**: **23.459** (el más bajo).  
- **Tienda 3**: **24.805**, también competitivo.  
- **Tienda 2**: **25.216**.  
- **Tienda 1**: **26.018**, el más alto.  

Los envíos más caros están en **Tienda 1**, mientras que Tienda 3 y Tienda 4 ofrecen mejores costos.  

---

## Conclusión y Recomendación  

Tras analizar todos los factores (ingresos, categorías más populares, calificaciones, productos vendidos y costes de envío), se concluye que:  

👉 **La tienda más recomendable para el Sr. Juan es la *Tienda 3*.**  

### ✅ Justificación  
- **Buenos ingresos**, en el tercer lugar pero muy cerca de Tienda 2.  
- **La mejor calificación promedio (4.05)** → refleja mayor satisfacción del cliente.  
- **Costo de envío competitivo (24.805)**, bastante más bajo que Tienda 1 y Tienda 2.  
- **Productos más vendidos alineados a las categorías fuertes** (muebles y electrodomésticos).  

Aunque **Tienda 1** tiene los mayores ingresos, su **baja calificación y alto costo de envío** la hacen menos atractiva.  
**Tienda 2** es sólida, pero no supera a Tienda 3 en satisfacción y costos.  
**Tienda 4**, aunque barata en envíos, genera los menores ingresos.  

📌 Por lo tanto, la **Tienda 3 ofrece el mejor equilibrio entre ventas, reputación y costos**, convirtiéndose en la opción más estratégica para que el Sr. Juan venda sus productos.
